## Simple code to show that the product of the lengths of chords to $n$ equally spaced points on a unit circle is equal to $n$. Inspired by this tweet: https://twitter.com/johncarlosbaez/status/1047136465690324992

```python
from complex import *
```

```python
unit = Complex(1)
```

```python
roots_5th = roots(5, unit)
```

```python
import matplotlib.pyplot as plt
from math import sqrt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

% matplotlib inline
```

```python
fig = plt.figure(figsize = (8,8))
plt.xticks([-1, 1])
plt.yticks([-1, 1])
ax = fig.add_subplot(1, 1, 1)

# I used this: https://stackoverflow.com/questions/2409774/how-can-i-produce-student-style-graphs-using-matplotlib
# for some of the formatting
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# set the background color so that it's not white
ax.set_facecolor('xkcd:light gray')

# draw the circle
circ = plt.Circle((0, 0), radius=1, edgecolor='black', facecolor='None')
ax.add_patch(circ)
plt.axis('scaled')


roots_rect = []
for root in roots_5th:
    z = rect(root[0], root[1])
    roots_rect.append(z)
    ax.plot(z.re, z.im, 'ro')

base = roots_rect[0]
lengths = []
for root in roots_rect[1:]:
    ax.plot([base.re, root.re], [base.im, root.im],'r')
    
    length = sqrt((base.re - root.re)**2 + (base.im - root.im)**2)
    lengths.append(length)
```

```python
from operator import mul
from functools import reduce
```

```python
prod = reduce(mul, lengths)

final_str = 'The product of the lengths of the chords is {:.2} and n is {}'.format(prod, 5)

print(final_str)
```
